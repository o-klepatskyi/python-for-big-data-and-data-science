# IV. Data Cleaning

11. (3б) Перетворити колонку launched у цілочислену (Int64). Перевірити, чи все пройшло, як очікувалось.

```python
missile_attacks['launched'] = missile_attacks['launched'].astype('Int64')
missile_attacks
```

```python
missile_attacks['launched'].dtype
```

12. (3б) Видалити колонки, де кількість непустих значень становить менше 30% від загальної кількості записів. Можна вручну порахувати і тут тільки видалити. За автоматизацію розрахунку відстоків - додаткові 3 бали.

```python
missile_attacks.dropna(axis=1, thresh=0.3 * missile_attacks.shape[0], inplace=True)
missile_attacks
```

13. (4б) Очистити датафрейм від дублікатів (дублюючихся рядків) та записів, де атака продовжувалась відʼємну або нульову кількість часу (що фізично неможливо). Не забудьте оновити індекс. Скільки рядків було видалено? Як ви це перевірили у коді?

```python
rows_before = missile_attacks.shape[0]

missile_attacks.drop_duplicates(inplace=True)

missile_attacks['time_start'] = pd.to_datetime(missile_attacks['time_start'], format="mixed")
missile_attacks['time_end'] = pd.to_datetime(missile_attacks['time_end'], format="mixed")

missile_attacks = missile_attacks[missile_attacks['time_end'] - missile_attacks['time_start'] > pd.Timedelta(0)]
missile_attacks.reset_index(drop=True, inplace=True)

rows_after = missile_attacks.shape[0]
print(f"Rows deleted: {rows_before - rows_after}")
```

17. (10б) Зʼясуйте, у який день було найбільше влучань за весь період (незбитих пристроїв). Виведіть час початку атаки та кількість влучань.

```python
missile_attacks.loc[:, "hits"] = (missile_attacks["launched"] - missile_attacks["destroyed"])
max_hit = missile_attacks["hits"].max()
day = missile_attacks[missile_attacks["hits"] == max_hit].iloc[0]
print(f"Start of the attack: {day['time_start']}, no of hits: {max_hit}")
```

```python
missile_attacks.loc[:, "duration"] = missile_attacks["time_end"] - missile_attacks["time_start"]

missile_attacks.loc[:, "year"] = missile_attacks["time_start"].dt.year
missile_attacks.loc[:, "month"] = missile_attacks["time_start"].dt.month

longest_attacks = missile_attacks.groupby(["year", "month"])["attack_duration"].max().reset_index(name="attack_duration").sort_values(by=["year", "month"], ascending=[False, True])
longest_attacks.to_csv("./data/longest_attacks.csv", index=False)
pd.read_csv("./data/longest_attacks.csv")
```
